```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los archivos CSV
df_mexico = pd.read_csv("Mexico.csv")
df_australia = pd.read_csv("Australia.csv")
df_buenos_aires = pd.read_csv("Buenos Aires.csv")
```

```python
#Verificamos información del DataFrame
df_mexico.info()
df_australia.info()
df_buenos_aires
```

```python
df_mexico.dropna(subset=['price', 'latitude', 'longitude'], inplace=True)
df_australia.dropna(subset=['price', 'latitude', 'longitude'], inplace=True)
df_buenos_aires.dropna(subset=['price', 'latitude', 'longitude'], inplace=True)
```

```python
# Imputación de valores faltantes con la mediana
df_mexico['reviews_per_month'].fillna(df_mexico['reviews_per_month'].median(), inplace=True)
df_mexico['number_of_reviews'].fillna(df_mexico['number_of_reviews'].median(), inplace=True)

df_australia['reviews_per_month'].fillna(df_australia['reviews_per_month'].median(), inplace=True)
df_australia['number_of_reviews'].fillna(df_australia['number_of_reviews'].median(), inplace=True)

df_buenos_aires['reviews_per_month'].fillna(df_buenos_aires['reviews_per_month'].median(), inplace=True)
df_buenos_aires['number_of_reviews'].fillna(df_buenos_aires['number_of_reviews'].median(), inplace=True)
```

```python
# Eliminar columnas con más del 50% de valores nulos
threshold = len(df_mexico) * 0.5
df_mexico.dropna(thresh=threshold, axis=1, inplace=True)

threshold = len(df_australia) * 0.5
df_australia.dropna(thresh=threshold, axis=1, inplace=True)

threshold = len(df_buenos_aires) * 0.5
df_buenos_aires.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
# Calcular los cuartiles y el IQR para la columna 'price'
Q1 = df_mexico['price'].quantile(0.25)
Q3 = df_mexico['price'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites superior e inferior
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtrar los outliers fuera de los límites
df_mexico = df_mexico[(df_mexico['price'] >= lower_limit) & (df_mexico['price'] <= upper_limit)]
```

```python
# Calcular la desviación estándar
mean_nights = df_mexico['minimum_nights'].mean()
std_nights = df_mexico['minimum_nights'].std()

# Definir los límites basados en la desviación estándar
upper_limit_nights = mean_nights + 3 * std_nights

# Reemplazar valores mayores al límite superior con el valor máximo permitido
df_mexico['minimum_nights'] = np.where(df_mexico['minimum_nights'] > upper_limit_nights, upper_limit_nights, df_mexico['minimum_nights'])
```

```python
# Calcular el percentil 99
percentile_99 = df_mexico['number_of_reviews'].quantile(0.99)

# Filtrar los outliers que exceden el percentil 99
df_mexico = df_mexico[df_mexico['number_of_reviews'] <= percentile_99]
```

```python
# Guardar el archivo CSV procesado
df_mexico.to_csv("Mexico_procesado.csv", index=False)
```

```python
def preprocesar(df_australia, Australia):
    print(f"Información de {Australia}:")
    df_australia.info()

    # Eliminar filas con valores nulos en 'price', 'latitude' y 'longitude'
    df_australia.dropna(subset=['price', 'latitude', 'longitude'], inplace=True)

    # Imputación de valores faltantes con la mediana
    df_australia['reviews_per_month'].fillna(df_australia['reviews_per_month'].median(), inplace=True)
    df_australia['number_of_reviews'].fillna(df_australia['number_of_reviews'].median(), inplace=True)

    # Eliminar columnas con más del 50% de valores nulos
    threshold = len(df_australia) * 0.5
    df_australia.dropna(thresh=threshold, axis=1, inplace=True)

    # Manejo de outliers usando IQR para 'price'
    Q1 = df_australia['price'].quantile(0.25)
    Q3 = df_australia['price'].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    df_australia = df_australia[(df_australia['price'] >= lower_limit) & (df_australia['price'] <= upper_limit)]

    # Manejo de outliers usando desviación estándar para 'minimum_nights'
    mean_nights = df_australia['minimum_nights'].mean()
    std_nights = df_australia['minimum_nights'].std()
    upper_limit_nights = mean_nights + 3 * std_nights
    df_australia['minimum_nights'] = np.where(df_australia['minimum_nights'] > upper_limit_nights, upper_limit_nights, df_australia['minimum_nights'])

    # Manejo de outliers usando percentil 99 para 'number_of_reviews'
    percentile_99 = df_australia['number_of_reviews'].quantile(0.99)
    df_australia = df_australia[df_australia['number_of_reviews'] <= percentile_99]

    # Guardar el archivo procesado
    df_australia.to_csv(f"{Australia}_procesado.csv", index=False)
    return df_australia

```

```python
def preprocesar(df_buenos_aires, ciudad):
    print(f"Información de {ciudad}:")
    df_buenos_aires.info()

    # Eliminar filas con valores nulos en 'price', 'latitude' y 'longitude'
    df_buenos_aires.dropna(subset=['price', 'latitude', 'longitude'], inplace=True)

    # Imputación de valores faltantes con la mediana
    df_buenos_aires['reviews_per_month'].fillna(df_buenos_aires['reviews_per_month'].median(), inplace=True)
    df_buenos_aires['number_of_reviews'].fillna(df_buenos_aires['number_of_reviews'].median(), inplace=True)

    # Eliminar columnas con más del 50% de valores nulos
    threshold = len(df_buenos_aires) * 0.5
    df_buenos_aires.dropna(thresh=threshold, axis=1, inplace=True)

    # Manejo de outliers usando IQR para 'price'
    Q1 = df_buenos_aires['price'].quantile(0.25)
    Q3 = df_buenos_aires['price'].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    df_buenos_aires = df_buenos_aires[(df_buenos_aires['price'] >= lower_limit) & (df_buenos_aires['price'] <= upper_limit)]

    # Manejo de outliers usando desviación estándar para 'minimum_nights'
    mean_nights = df_buenos_aires['minimum_nights'].mean()
    std_nights = df_buenos_aires['minimum_nights'].std()
    upper_limit_nights = mean_nights + 3 * std_nights
    df_buenos_aires['minimum_nights'] = np.where(df_buenos_aires['minimum_nights'] > upper_limit_nights, upper_limit_nights, df_buenos_aires['minimum_nights'])

    # Manejo de outliers usando percentil 99 para 'number_of_reviews'
    percentile_99 = df_buenos_aires['number_of_reviews'].quantile(0.99)
    df_buenos_aires = df_buenos_aires[df_buenos_aires['number_of_reviews'] <= percentile_99]

    # Guardar el archivo procesado
    df_buenos_aires.to_csv(f"{ciudad}_procesado.csv", index=False)

    return df_buenos_aires

# Cargar los archivos CSV
df_mexico = pd.read_csv("Mexico.csv")
df_australia = pd.read_csv("Australia.csv")
df_buenos_aires = pd.read_csv("Buenos Aires.csv")

# Aplicar pre-procesamiento a los tres DataFrames
df_mexico = preprocesar(df_mexico, "Mexico")
df_australia = preprocesar(df_australia, "Australia")
df_buenos_aires = preprocesar(df_buenos_aires, "Buenos_Aires")
```
